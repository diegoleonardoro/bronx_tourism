```python
import pandas as pd
```

```python
data = pd.read_csv ("data/AB_NYC_2019.csv")
```

```python
bronx = data [data["neighbourhood_group"]=="Bronx"]
```

```python
bronx ["last_review"].unique()
```

```python
bronx= bronx [bronx ["last_review"]!= "nan" ]
```

```python
bronx["year"] = bronx["last_review"].str[:4]
```

```python
bronx = bronx [bronx ["year"]!="NaN"]
```

```python
years = bronx["year"].unique()
```

```python
airbnb_avgs = {}
for year in years:
    x = bronx [bronx["year"]== str (year)]
    try:
        avg = sum( x["price"]) / len (x["price"])
        
    except ZeroDivisionError:
        continue
    airbnb_avgs[year] = avg
    
```
